```r
# import library for %>%
# library(tidyverse)

# load libraries
# library(tidyverse)
library(ggplot2)

# load data
csgo_round_snapshots <- read.csv("csgo_round_snapshots.csv")

# print data types
# str(csgo_round_snapshots[, -c(4, 5, 97)])

# make all constant zero columns to constnt 1
csgo_round_snapshots <- csgo_round_snapshots[, -c(4, 5, 97)]


constant_columns <- c()
for (i in 1:ncol(csgo_round_snapshots)) {
  if (length(unique(csgo_round_snapshots[, i])) == 1) {
    constant_columns <- c(constant_columns, i)
  }
}

# constant_columns


# make pca plot 
pca <- prcomp(csgo_round_snapshots[, -constant_columns], scale = TRUE)
pca$round_winner <- csgo_round_snapshots$round_winner
# plot the pca
ggplot(as.data.frame(pca$x), aes(x = PC1, y = PC2)) + geom_point()
```
